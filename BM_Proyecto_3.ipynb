```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def functionGenerator(par):
  def temp(x, ti):
    return x*(1-x)*(x/par[0] -1)
  return temp

ti = np.linspace(0,.1,100)
a=0.002
par=[a]
allee = functionGenerator(par)

x0=0.0005
sol = odeint(allee, x0, ti)
plt.figure(figsize=(10, 6))
plt.axhline(0, color="black", ls='--',lw=1, label= r'raíces($0,\alpha,1$)')
plt.axhline(a, color="black", ls='--',lw=1)
plt.axhline(1, color="black",ls='--',lw=1)
plt.plot(ti, sol, label='$x_0=0.0005$')

x0=0.04
sol = odeint(allee, x0, ti)
plt.plot(ti, sol, label='$x_0=0.04$')

x0=0.1
sol = odeint(allee, x0, ti)
plt.plot(ti, sol, label='$x_0=0.1$')

x0=0.4
sol = odeint(allee, x0, ti)
plt.plot(ti, sol, label='$x_0=0.4$')

x0=0.6
sol = odeint(allee, x0, ti)
plt.plot(ti, sol, label='$x_0=0.6$')

x0=1.2
sol = odeint(allee, x0, ti)
plt.plot(ti, sol, label='$x_0=1.2$')

x0=2
sol = odeint(allee, x0, ti)
plt.plot(ti, sol, label='$x_0=2.0$')

x0=3
sol = odeint(allee, x0, ti)
plt.plot(ti, sol, label='$x_0=3.0$')

plt.ylabel('Relación Poblacional x* ($N/k$)')
plt.xlabel("Relación temporal ($rt$)")
plt.title(r'Análisis Dinámico con $\alpha=0.002$ [$A/k$]' )

plt.legend()
plt.show()
```
