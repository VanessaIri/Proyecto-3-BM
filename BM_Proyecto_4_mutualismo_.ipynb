```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def functionGenerator(par):
  def temp(p,ti):
    x,y=p
    dx=x*(1-x+par[0]*y)
    dy=par[2]*y*(1-y+par[1]*x)
    return dx,dy
  return temp

#parametros version ax = ay <1 b=1
a_x=0.5      #nfluencia de x en y
a_y=0.5      #influencia de y en x
b=1          #relacion tasas de crecimiento y/x
par=[a_y,a_x,b]

mutualism=functionGenerator(par)

ti = np.linspace(0, 100, 20000)

#Puntos fijos (0,0) (0,1) (1,0) [(1+ay)/1-axay, (1+ax)/1-axay]
p0=[0.001,0.001]
sol=odeint(mutualism,p0,ti)
plt.figure(figsize=(10, 6))
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.01,0.001]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.3,0.001]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.9,0.001]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[1.1,0.001]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[1.6,0.001]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[1.9,0.001]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")


p0=[0.01,1.9]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.3,1.9]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.9,1.9]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[1.1,1.9]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[1.6,1.9]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[1.9,1.9]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

#Cambios en y
p0=[0.001,0.01]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.001,0.3]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.001,0.9]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.001,1.1]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.001,1.6]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[0.001,1.9]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")


p0=[1.9,0.3]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[1.9,0.9]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[1.9,1.1]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")

p0=[1.9,1.6]
sol=odeint(mutualism,p0,ti)
plt.plot(sol[:, 0], sol[:, 1],color="skyblue")


plt.plot([0,0,1,(1 + a_y)/(1 - a_y*a_x)],[0,1,0,(1 +a_x)/(1 - a_y*a_x)],'o',color="darkred")
plt.ylabel(r"Poblacion Y ($y/k_y$)")
plt.xlabel(r"Poblacion X ($x/k_x$)")
plt.title(r'Espacio Fase con $\alpha_x=\alpha_y<1$ y  $\beta=1$' )
plt.show()


```
